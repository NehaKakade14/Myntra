```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Myntra Dataset.csv')

# Display the first few rows to verify
print(df.head())
```

```python
import ast

# Parse the JSON strings in the 'p_attributes' column, handling NaN values
df['p_attributes'] = df['p_attributes'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else {})

# Display the first few rows to verify
print(df.head())
```

```python
# Create new columns from the keys in the JSON strings
attributes_df = df['p_attributes'].apply(pd.Series)

# Display the first few rows of the new dataframe
print(attributes_df.head())
```

```python
# Combine the new columns with the original dataframe
df = pd.concat([df, attributes_df], axis=1)

# Drop the original 'p_attributes' column if no longer needed
df.drop(columns=['p_attributes'], inplace=True)

# Display the first few rows of the updated dataframe
print(df.head())

# Save the updated dataframe to a new CSV file
df.to_csv('MyntraDatasetNew.csv', index=False)
```
